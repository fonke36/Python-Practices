# Use of (**) operator in dictionary and function

```python
# Merging two dictionaries

Default = {'theme': "Dark", 'brightness': 30}
user_set = {'night_mode': True}

setting = {**Default, **user_set}

print(setting)
```

```python
# Updating dictionary

user_set_2 = {'brightness': 70, 'night_mode': False}

setting = {**setting, **user_set_2}

print(setting)
```

```python
# Left side and right side

setting = {**user_set_2, **setting}

print(setting)
```
